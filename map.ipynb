```python
class Graph:
    def __init__(self,ports,courses):
        self.ports = ports
        self.courses = courses
        
    def get_fee(self,name):
        for pt in self.ports:
            if pt.name == name:
                if pt.is_free == True:
                    return 0
                return pt.fee
        return 0
        
    def get_distance(self,start,end):
        for course in self.courses:
            i = course.start.name
            j = course.end.name
            k = course.distance
            if i == start and j == end:
                return k
            elif i == end and j == start:
                return k
        return 0
        
    def find_neighbors(self,name):
        ports = {}
        for course in self.courses:
            #print(course.start)
            if course.start.name == name and course.start.built == True:
                neighbor = course.end
                ports[neighbor] = 1
            elif course.end.name == name and course.end.built == True:
                neighbor = course.start
                ports[neighbor] = 1
        return list(ports)

    def open_port(self,names):
        for name in names:
            for pt in self.ports:
                if pt.name == name:
                    pt.is_free = True

    def close_port(self,names):
        for name in names:
            for pt in self.ports:
                if pt.name == name:
                    pt.is_free = False

    def fewest_ports(self,start,end): # more time at sea, BFS
        visited = []
        parents = {}
        frontier = [start]
        while frontier != []:
            cur = frontier[0]
            if cur == end:
                break
            frontier = frontier[1:]
            visited += [cur]
            #print(self.find_neighbors(cur))
            for neighbor in self.find_neighbors(cur):
                if neighbor.name not in visited and neighbor.name not in frontier:
                    frontier += [neighbor.name]
                    parents[neighbor.name] = cur
        # trace back
        #print(visited)
        distance = 0
        cost = 0
        path = [end]
        c = end
        p = parents[path[0]]
        max_iter = 500
        for _ in range(max_iter):
            cost += self.get_fee(c)
            distance += self.get_distance(c,p)
            path = [c] + path[:]
            c = p
            if c == start:
                path = [start] + path[:]
                break
            p = parents[c]
        return path[:-1], cost, distance

    def shortest_distance(self,start,end):# fatest way
        visited = []
        dist = {port.name:100000 for port in self.ports}
        dist[start] = 0
        frontier = [(0,start)]
        parents = {}
        while frontier != []:
            cur = frontier[0]
            port = cur[1]
            visited += [port]
            frontier = frontier[1:]
            for neighbor in self.find_neighbors(cur[1]):
                near = neighbor.name
                dis = self.get_distance(port, near)
                # print(dis)
                if (dis,near) not in frontier and near not in visited:
                    frontier += [(dis,near)]
                    frontier = sorted(frontier)
                if dist[near] > dist[port] + dis:
                    dist[near] = dist[port] + dis
                    parents[near] = port
        # trace back
        # print(dist)
        distance = 0
        cost = 0
        path = [end]
        c = end
        p = parents[path[0]]
        max_iter = 500
        for _ in range(max_iter):
            cost += self.get_fee(c)
            distance += self.get_distance(c,p)
            path = [c] + path[:]
            c = p
            if c == start:
                path = [start] + path[:]
                break
            p = parents[c]
        return path[:-1], cost, distance
                
    def least_cost(self,start,end): # least money pay to the ports
        visited = []
        dist = {port.name:100000 for port in self.ports}
        dist[start] = 0
        frontier = [(0,start)]
        parents = {}
        while frontier != []:
            cur = frontier[0]
            port = cur[1]
            visited += [port]
            frontier = frontier[1:]
            for neighbor in self.find_neighbors(cur[1]):
                near = neighbor.name
                cost = self.get_fee(near)
                # print(dis)
                if (cost,near) not in frontier and near not in visited:
                    frontier += [(cost,near)]
                    frontier = sorted(frontier)
                if dist[near] > dist[port] + cost:
                    dist[near] = dist[port] + cost
                    parents[near] = port
        # trace back
        # print(dist)
        distance = 0
        cost = 0
        path = [end]
        c = end
        p = parents[path[0]]
        max_iter = 500
        for _ in range(max_iter):
            cost += self.get_fee(c)
            distance += self.get_distance(c,p)
            path = [c] + path[:]
            c = p
            if c == start:
                path = [start] + path[:]
                break
            p = parents[c]
        return path[:-1], cost, distance

    def most_pirate_ports(self): # stay in pirate ports
        pass

    def most_my_ports(self): #stay in my port
        pass

    def trade_travel(self,start,end,op='cost'):
        if op == 'fewest':
            return self.fewest_ports(start,end)
        elif op == 'shortest':
            return self.shortest_distance(start,end)
        else:
            return self.least_cost(start,end)
```

```python
def round_value(money):
    if money % 10 > 5:
        money = money // 1
        while money % 10 != 0:
            money += 1
        return money
    if money % 5 > 3:
        money = money // 1
        while money % 5 != 0:
            money += 1
        return money
    else:
        money = money // 1
        while money % 5 != 0:
            money -= 1
        return money
print(round_value(money))
```
